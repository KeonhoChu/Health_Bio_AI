# 2. 모멘트 계산을 이용한 회전된 각도 계산
: 어떤 이미지가 원본 이미지로부터 theta 만큼 회전 되었을 때, Central moment를 가지고 물체의 회전된 각도를 파악할 수 있다는 원리

## 2.1. 기하학적 모멘트 함수 정의: 영상이 이동하면 그값이 변함

```python
def calculateMpq(img, p, q):
    mpq = 0
    for y in range(img.shape[0]):
        for x in range(img.shape[1]):
            mpq += np.power(x, p) * np.power(y, q) * img[y, x]
    return mpq
```

## 2.2. 중심 모멘트 정의: 영상의 무게 중심을 고려하여 모멘트를 계산

```python
def calculateMupq(img, p, q, x0, y0, m00):
    mupq= 0
    for y in range(img.shape[0]):
        for x in range(img.shape[1]):
            mupq += np.power((x - x0), p) * np.power((y - y0), q) * img[y, x]
    mupq = mupq / m00;
    return mupq
```

```python
m00 = calculateMpq(img, 0, 0)
m10 = calculateMpq(img, 1, 0)
m01 = calculateMpq(img, 0, 1)

x0 = m10 / m00
y0 = m01 / m00
```

```python
u11 = calculateMupq(img, 1, 1, x0, y0, m00);
u20 = calculateMupq(img, 2, 0, x0, y0, m00);
u02 = calculateMupq(img,0, 2, x0, y0, m00);
```

```python
theta = np.arctan(2*u11/(u20-u02))/2
angle = theta * 180 /np.pi
print(f"회전된 각도: {angle}")
```

```python
M = cv2.getRotationMatrix2D((img.shape[1]//2, img.shape[0]//2), angle, 1.0)
rotated_img = cv2.warpAffine(img, M, (img.shape[1], img.shape[0]))
plt.figure(figsize=(5,5))
plt.imshow(rotated_img, cmap=plt.cm.bone)
plt.show()
```
